#### Importing Packages

````python
import numpy as np
import pandas as pd

import yfinance as yf #Library to live maarket data pull

import plotly.graph_objs as go #Ploty visualization
````

#### Define parameters for live data

````python
#Interval required 5 minutes
data = yf.download(tickers='PYPL', period='30d', interval='1d')
#Print data
data.head()
````

````python
data['Close'].mean()
````

#### Plotting a graph using Plotly

````python
fig = go.Figure()

fig.add_trace(go.Candlestick(x=data.index,
                open=data['Open'],
                high=data['High'],
                low=data['Low'],
                close=data['Close'], name = 'market data'))

fig.update_layout(
    title='PayPal Live Price Monitoring tool',
    yaxis_title='Stock Price (USD per Shares)')

fig.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=15, label="15m", step="minute", stepmode="backward"),
            dict(count=45, label="45m", step="minute", stepmode="backward"),
            dict(count=1, label="HTD", step="hour", stepmode="todate"),
            dict(count=3, label="3h", step="hour", stepmode="backward"),
            dict(step="all")
        ])
    )
)

fig.show()
````
